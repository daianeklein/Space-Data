# AVERAGE AGE OF COUNTRIES

***

Average age of different countries starting from 1950 to prediction till 2050.

**About this file**

This file contains the median ages of 215 different regions of the world, some of which are not sovereign states as well. The ages are calculated every 5 years from 1950 till 2050 in the future. One thing to note however is the ages in the future are a prediction and obviously not calculated.

Dataset: https://www.kaggle.com/divyansh22/average-age-of-countries

# Lib's

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime 

import requests

import warnings
warnings.filterwarnings('ignore')
```

# DATASET

```python
df = pd.read_csv(r'/Users/daianeklein/Downloads/MedianAge.csv')
df
```

```python
# Paises
df['Country'].unique()
```

```python
#melt

df = df.melt('Country', var_name='year', value_name='idade_media')
df.head(5)
```

# DATA VIZ

```python
# geral

plt.figure(figsize=(16,3))
ax = sns.lineplot(data = df.groupby(['year'], as_index=False).mean(), x ='year', y='idade_media')
sns.despine(bottom=True, left=True)
plt.title('Idade média da população mundial ao longo dos anos', pad = 20)
plt.xlabel('Year', labelpad=20)
plt.xlabel('Idade Média', labelpad=20)
ax.grid(axis='y', linewidth=0.5, color='lightgrey')

ax;
```

### HISTOGRAMA

```python
plt.figure(figsize=(16,5))

sns.histplot(data = df,    bins = 20, binwidth=1, linewidth = 0.3, palette = 'mako', )
sns.despine(bottom=True, left=True)
plt.title('Histograma - Idade média da população mundial ao longo dos anos', pad = 20)
plt.xlabel('Year', labelpad=20)
plt.xlabel('Idade Média', labelpad=20)

ax;
```

```python
df['year'].unique()
```

```python
# Slice dos anos
ano1 = df['year'].unique()[:7] #até o ano 1980
ano2 = df['year'].unique()[7:15] #do ano 1985 até 2020
ano3 = df['year'].unique()[15:] # de 2025 até 2050
```

```python
plt.figure(figsize=(16, 5))

ax1 = plt.subplot(1, 3, 1)
sns.histplot(data = df.query('year in @ano1'),   bins = 5, binwidth=2, linewidth = 0.3, palette = 'mako')
sns.despine(bottom=True, left=True)
plt.title('Até o ano 1980', pad = 20)
plt.xlabel('Idade média', labelpad=20)
plt.ylabel([],fontsize=0)
plt.legend([],[], frameon=False)

ax2 = plt.subplot(1, 3, 2)
sns.histplot(data = df.query('year in @ano2'),   bins = 5, binwidth=2, linewidth = 0.3, palette = 'mako')
sns.despine(bottom=True, left=True)
plt.title('De ano 1985 até 2020', pad = 20)
plt.xlabel('idade média', labelpad=20)
plt.ylabel([],fontsize=0)
plt.legend([],[], frameon=False)


ax3= plt.subplot(1, 3, 3)
sns.histplot(data = df.query('year in @ano3'),   bins = 5, binwidth=2, linewidth = 0.3, palette = 'mako')
sns.despine(bottom=True, left=True)
plt.title('Do ano 2025 até 2050', pad = 20)
plt.xlabel('Idade média', labelpad=20)
plt.ylabel([],fontsize=0)
plt.legend([],[], frameon=False)

ax3;
```

### POR PAÍS

```python
paises = ['Brazil','United States','France','Japan','China','South Africa','New Zealand','Bangladesh','Greece','Switzerland','Syria','Chile']
```

```python
df.query('Country in @paises')
```

```python
plt.figure(figsize=(16,5))


sns.boxplot(data = df.query('Country in @paises').sort_values(by='idade_media') , y='idade_media', x='Country',
           color = 'steelblue', linewidth = 0.5)

sns.despine(bottom=True, left=True)
plt.title('Boxplot - Idade média', pad = 20)
plt.xlabel('Países', labelpad=20)
plt.ylabel('Idade média', labelpad=20)
plt.ylabel([],fontsize=0)
```

### VISUALIZANDO POR PAÍS 

```python
for p in paises:
    
    plt.figure(figsize=(8,4))

    sns.histplot(data = df[df['Country'] == p],   bins = 5, kde = True, binwidth=2, linewidth = 0.3, palette = 'mako')
    sns.despine(bottom=True, left=True)
    plt.title('Histograma - Idade média da população no(a) {}'.format(p), pad = 20)
    plt.xlabel('Year', labelpad=20)
    plt.xlabel('Idade Média', labelpad=20)

    ax;
```

### DATA VIZ - PAISES POR ANO

```python
for p in paises:

    plt.figure(figsize=(16, 5))

    ax1 = plt.subplot(1, 3, 1)

    sns.histplot(data = df[(df['Country'] == p) & (df['year'].isin(ano1))],   bins = 5, kde = True,
                 linewidth = 0.5, palette = 'mako')
    sns.despine(bottom=True, left=True)
    plt.title('Idade média da população em {} até 1980'.format(p), pad = 20)
    plt.xlabel('Year', labelpad=20)
    plt.xlabel('Idade Média', labelpad=20)
    plt.legend([],[], frameon=False)
    plt.ylabel([],fontsize=0)

    ax2 = plt.subplot(1, 3, 2)
    sns.histplot(data = df[(df['Country'] == p) & (df['year'].isin(ano2))],   bins = 5, kde = True,
                linewidth = 0.3, palette = 'mako')
    sns.despine(bottom=True, left=True)
    plt.title('De 1985 à 2020', pad = 20)
    plt.xlabel('Year', labelpad=20)
    plt.xlabel('Idade Média', labelpad=20)
    plt.legend([],[], frameon=False)
    plt.ylabel([],fontsize=0)

    ax3= plt.subplot(1, 3, 3)
    sns.histplot(data = df[(df['Country'] == p) & (df['year'].isin(ano3))],   bins = 5, kde = True,
                linewidth = 0.3, palette = 'mako')
    sns.despine(bottom=True, left=True)
    plt.title('De 2025 à 2050', pad = 20)
    plt.xlabel('Year', labelpad=20)
    plt.xlabel('Idade Média', labelpad=20)
    plt.legend([],[], frameon=False)
    plt.ylabel([],fontsize=0)

    ax3;
```
